```python
import numpy as np
x = range(1,7)
x = list(x)
acc_per_epoch = np.array([0.8345, 0.8472, 0.8351, 0.8355, 0.8353, 0.8357])
loss_per_epoch = np.array([0.6626, 0.6315, 0.6088, 0.5648, 0.5229, 0.8472])
```

```python
import matplotlib.pyplot as plt
```

```python

plt.xlabel('Epoch')
plt.ylabel('Acc , Loss')
plt.plot(x, acc_per_epoch, 'g--', x, loss_per_epoch, 'r--')
```
